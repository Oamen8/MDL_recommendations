```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
```

```python
user_id = 'Oamen'
```

```python
user_item_matrix = pd.read_csv('pivoted.csv',index_col=0).fillna(0)
user_item_matrix
```

```python
user_similarity = cosine_similarity(user_item_matrix)
user_similarity_df = pd.DataFrame(user_similarity, index=user_item_matrix.index, columns=user_item_matrix.index)
user_similarity_df
```

```python
def get_similar_users(user_id, n=5):
    similar_users = user_similarity_df.loc[user_id].sort_values(ascending=False)[1:n+1]
    return similar_users
```

```python
def predict_rating(user_id, item_id, n_similar_users=5):
    similar_users = get_similar_users(user_id, n_similar_users)
    user_ratings = user_item_matrix.loc[similar_users.index, item_id]
    weighted_ratings = (user_ratings * similar_users.values).sum() / similar_users.sum()
    # print(similar_users)
    return weighted_ratings

def get_recommendations(user_id, n_recommendations=20):
    user_ratings = user_item_matrix.loc[user_id]
    unrated_items = user_ratings[user_ratings == 0].index
    predicted_ratings = [predict_rating(user_id, item) for item in unrated_items]
    recommendations = pd.Series(predicted_ratings, index=unrated_items).sort_values(ascending=False)
    return recommendations.head(n_recommendations)
```

```python
similar_users = get_similar_users(user_id, 20)

plt.stem(similar_users.index,similar_users,orientation='horizontal')

plt.gca().invert_yaxis()

plt.xlabel('similarity')
plt.ylabel('user')
plt.title('20 Most Similar Users to '+user_id)

plt.show()
```

```python
recommendations = get_recommendations(user_id)
print(f"Top 20 recommendations for user {user_id}:")
print(recommendations)
```

```python
item_features = pd.read_csv('item_features.csv',index_col=0)
item_features_encoded = pd.get_dummies(item_features, columns=['actor1','actor2','genre1','genre2','country'])
item_features_encoded
```

```python
user_preferences = user_item_matrix.dot(item_features_encoded)
# user_preferences_normalized = user_preferences.div(user_preferences.sum(axis=1), axis=0)

# Add a small epsilon to avoid division by zero
epsilon = 1e-10
user_preferences_sum = user_preferences.sum(axis=1) + epsilon

# Normalize user preferences, handling zero rows
user_preferences_normalized = user_preferences.div(user_preferences_sum, axis=0)
user_preferences_normalized.loc[user_id]
```

```python
user_preferences_normalized.to_csv('user_pref.csv')
```

```python
actor_series = user_preferences_normalized.loc[:, user_preferences_normalized.columns.str.startswith('actor')].loc[user_id]
actor_series1 = user_preferences_normalized.loc[:, user_preferences_normalized.columns.str.startswith('actor1')].loc[user_id]
actor_series2 = user_preferences_normalized.loc[:, user_preferences_normalized.columns.str.startswith('actor2')].loc[user_id]

genre_series = user_preferences_normalized.loc[:, user_preferences_normalized.columns.str.startswith('genre')].loc[user_id]
country_series = user_preferences_normalized.loc[:, user_preferences_normalized.columns.str.startswith('country')].loc[user_id]

actor_series.index = actor_series.index.str.removeprefix('actor1_').str.removeprefix('actor2_')
actor_series1.index = actor_series1.index.str.removeprefix('actor1_')
actor_series2.index = actor_series2.index.str.removeprefix('actor2_')

genre_series.index = genre_series.index.str.removeprefix('genre1_').str.removeprefix('genre2_')
country_series.index = country_series.index.str.removeprefix('country_')
```

```python
top_actors = pd.to_numeric(actor_series).nlargest(25)

plt.stem(top_actors.index,top_actors,orientation='horizontal')

plt.gca().invert_yaxis()

plt.xlabel('Normalized Feature Importance')
plt.ylabel('Leads')
plt.title('Top 25 Leads for '+user_id)

plt.show()

# Two points for Han So Hee indicating for when she is listed as the first or second actor (first is larger value point)
```

```python
genre_ordered = genre_series.sort_values(ascending=False)

plt.stem(genre_ordered.index,genre_ordered,orientation='horizontal')

plt.gca().invert_yaxis()

plt.xlabel('Normalized Feature Importance')
plt.ylabel('Genre')
plt.title('Top Genres for '+user_id)

plt.show()

# Two points for each genre representing their listing as the first or second genre
# Interestingly Action is the most common, and never listed second.
```

```python
country_ordered = country_series.sort_values(ascending=False)[:4]

# plt.stem(country_ordered.index,country_ordered,orientation='horizontal')

# plt.gca().invert_yaxis()

def format_proportion(value):
    return f'{value:.1f}'  # Format value with one decimal place

plt.pie(country_ordered,labels=['South Korea', 'Thailand','Japan','China'],autopct=lambda p: format_proportion(p),startangle=15,pctdistance=1.2,labeldistance=1.4)
plt.title('Top Countries for '+user_id)

plt.show()
```

```python
def recommend_items(user_id, user_item_matrix, user_preferences_normalized, item_features, n_recommendations=5):
    user_ratings = user_item_matrix.loc[user_id]
    unrated_items = user_ratings[user_ratings == 0].index
    
    predicted_ratings = {}
    for item_id in unrated_items:
        item_feature_vector = item_features.loc[item_id]
        user_preference_vector = user_preferences_normalized.loc[user_id]
        
        # Calculate similarity score
        predicted_rating = user_preference_vector.dot(item_feature_vector)
        predicted_ratings[item_id] = predicted_rating
    
    recommended_items = sorted(predicted_ratings.items(), key=lambda x: x[1], reverse=True)
    return recommended_items[:n_recommendations]
```

```python
recommendations = recommend_items(user_id=user_id, 
                                   user_item_matrix=user_item_matrix, 
                                   user_preferences_normalized=user_preferences_normalized, 
                                   item_features=item_features_encoded, 
                                   n_recommendations=20)
print(recommendations[:10])
```

```python
x,y = zip(*recommendations)
plt.stem(x,y,orientation='horizontal')

plt.gca().invert_yaxis()

plt.xlabel('Predicted Preference Score') # not the predicted rating out of 10, but the probability the user will 'like' it
plt.ylabel('Title')
plt.title('Top 20 Recommendations for '+user_id)

plt.show()
```
